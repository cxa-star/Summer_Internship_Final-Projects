```python
# Business Rule 1:
# Employees ko salary ke basis par categories me divide karna.
# Isse pata chalega kaun High Salary, Medium Salary aur Low Salary group me hai.

def salary_category(salary):

    if salary >= 100000:
        return "High Salary"

    elif salary >= 70000:
        return "Medium Salary"

    else:
        return "Low Salary"

# Function ko Monthly_Salary column par apply karke
# ek naya column Salary_Category create kar rahe hain
df["Salary_Category"] = df["Monthly_Salary"].apply(salary_category)


# Salary Category business rule isliye apply kiya gaya taaki
# employees ko salary ranges ke according classify kiya ja sake aur salary distribution ko easily analyze kiya ja sake
```

```python
# Business Rule 2:
# Employees ko experience ke basis par classify karna.
# Isse pata chalega employee Junior, Mid-Level ya Senior hai.

def experience_level(exp):
    if exp <= 2:
        return "Junior"

    elif exp <= 7:
        return "Mid-Level"

    else:
        return "Senior"

# Function ko Experience_Years column par apply karke
# ek naya column Experience_Level create kar rahe hain
df["Experience_Level"] = df["Experience_Years"].apply(experience_level)


# Experience Level business rule ka purpose employees ko unke experience ke according categorize karna tha,
# jisse organization me Junior, Mid-Level aur Senior workforce ka analysis kiya ja sake."
```

```python
print("Average Salary Of Employee :", df["Monthly_Salary"].mean())
```

```python
# idxmax() -  highest salary wali row ka index find karta hai
# .loc[] -  us row ki complete details retrieve karta hai."

print("Highest Salary Employee :", df.loc[df["Monthly_Salary"].idxmax()])
```

```python
# Ye salary distribution ko department basis par show kar rha hai.

print("Hghest salary on the basis of", df.groupby("Department")["Monthly_Salary"].mean())
```

```python
# Ye salary distribution ko gender basis par show kar rha h

print("Highest salary on the bais of", df.groupby("Gender")["Monthly_Salary"].mean())
```
