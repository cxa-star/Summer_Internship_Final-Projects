```python
plt.figure(figsize=(3,3))
sns.scatterplot(x='Customer ID', y='Purchase Amount (USD)', data=df)
plt.show()
```

```python
df.drop('Customer ID', axis=1, inplace=True)
```

```python
# Numeric vs Numeric   | Scatter Plot 
# Category vs Numeric  | Bar Plot   
# Category vs Category | Count Plot   
# Numeric Distribution | Histogram    
# Outliers             | Box Plot     
```

```python
plt.figure(figsize=(3,3))
sns.scatterplot(x='Age', y='Purchase Amount (USD)', data=df)
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.scatterplot(x='Review Rating', y='Purchase Amount (USD)', data=df)
plt.show()
```

```python
plt.figure(figsize=(3,3))
sns.scatterplot(x='Previous Purchases', y='Purchase Amount (USD)', data=df)
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(df.select_dtypes(include=['int64','float64']).corr(), annot=True, cmap='coolwarm')
plt.show()

# I used a heatmap to understand the correlation between numerical features and 
# identify whether any numeric variables have a strong relationship with Purchase Amount (USD).
```

```python
# Kaunsi product category sabse zyada revenue generate karti hai?
sns.barplot(x='Category',y='Purchase Amount (USD)',data=df)
plt.show() 
```

```python
# Kya subscribers non-subscribers se zyada spend karte hain?
sns.barplot(x='Subscription Status',y='Purchase Amount (USD)',data=df)
plt.show()
```

```python
# Kya discount dene se spending badhti hai?
sns.barplot(x='Discount Applied',y='Purchase Amount (USD)',data=df)
plt.show()
```

```python
# Kis season me customers sabse zyada spend karte hain?
sns.barplot(x='Season',y='Purchase Amount (USD)',data=df)
plt.show()
```

```python
# The graphs were created to understand how different customer and
# business factors such as category, subscription status, discounts, and season influence customer spending behavior.
```

```python
# Customer spending remains relatively consistent across different categories, seasons, subscription statuses, and discount conditions.
# No single factor shows a strong impact on purchase amount. 
# However, these variables are still useful for understanding customer behavior and can be explored further in future machine learning projects.
```

```python
sns.histplot(df['Purchase Amount (USD)'], kde=True)
plt.show()           
# To understand the distribution of customer spending.
```

```python
sns.countplot(x='Category', data=df)
plt.show()
# Kaunsi category me kitne purchases hue.
```

```python
sns.countplot(x='Season', data=df)
plt.show()
# Kis season me kitne transactions hue.
```

```python
sns.boxplot(y='Purchase Amount (USD)', data=df)
plt.show()
# to check outliers
```
