# Hypothesis

```python
# Hypothesis = Ek assumption (maan lena)

# example :- Airline company sochti hai ---> Loyal customers zyada distance travel karte honge.
# Ye bas ek assumption hai.
# Statistics ka kaam hai ---> "Prove karo data ke basis par."

# Hypothesis Hamesha 2 Hoti Hain --> 1. H₀ (Null Hypothesis) --> Ye bolti hai--- Koi difference nahi hai.
# Example ---> Loyal aur Disloyal customers ki average flight distance same hai.

# 2. H₁ (Alternative Hypothesis) ---> Ye bolti hai --- Difference hai.
# Example ---> Loyal aur Disloyal customers ki average flight distance different hai.


# Statistical Test ---> suppose hypothesis likh di, ab usko prove kaise karnege... tb statistical test krte hai..

# age 2 groups h to T-test , or agr 3 ya 3+ groups h to... ANOVA test ( Analysis of Variance ) ---> Ye ek statistical test hai jo check karta hai -
# "3 ya usse zyada groups ke means (average values) same hain ya different?

# Agar Dono Columns Categorical Hain --> Chi-Square test use karenge
# Agar Dono Numerical Hain --> Correlation
# 
```

# T-Test :-

```python
# Loyal customers ki flight distance
loyal = df[df["Customer Type"]=="Loyal Customer"]["Flight Distance"]

# Disloyal customers ki flight distance
disloyal = df[df["Customer Type"]=="disloyal Customer"]["Flight Distance"]

# T-Test lagana
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(loyal, disloyal)
print("P-value =", p_value)
```

```python
print("{:.20e}".format(p_value))
```

# Interpretation

```python
# Since p-value is less than 0.05,the null hypothesis is rejected.
# There is a statistically significant difference between Loyal and Disloyal customers' average flight distance.
```

```python
if p_value < 0.05:

    print("Reject H0")
    print("Travel Class significantly affects Flight Distance")

else:
    print("Fail to Reject H0")
    print("Travel Class does not significantly affect Flight Distance")
```

# ANOVA :-

```python
eco = df[df["Class"]=="Eco"]["Flight Distance"]

business = df[df["Class"]=="Business"]["Flight Distance"]

eco_plus = df[df["Class"]=="Eco Plus"]["Flight Distance"]

from scipy.stats import f_oneway
f_stat, p_value = f_oneway(eco,business,eco_plus)
print(p_value)
```

# Interpretation 

```python
if p_value < 0.05:

    print("Reject H0")
    print("Travel Class significantly affects Flight Distance")

else:
    print("Fail to Reject H0")
    print("Travel Class does not significantly affect Flight Distance")
```

# Chi-Square Test :-

```python
# Business Question :- Is Customer Type related to Satisfaction?

# Hypothesis :- H₀ --> Customer Type and Satisfaction are independent.

# H₁ --> Customer Type and Satisfaction are related.
```

```python
table = pd.crosstab(
    df["Customer Type"],
    df["satisfaction"]
)

from scipy.stats import chi2_contingency
chi2, p_value, dof, expected = chi2_contingency(table)
print(p_value)
```

```python
if p_value < 0.05:
    print("Reject H0")
    print("Customer Type and Satisfaction are related.")

else:
    print("Fail to Reject H0")
    print("Customer Type and Satisfaction are independent.")
```

# Correlation :-

```python
# Business Question :- Are Departure Delay and Arrival Delay related?

# Hypothesis ---> H₀ :- No relationship exists....... H₁ :- Relationship exists.
```

```python
from scipy.stats import pearsonr

corr, p_value = pearsonr(
    df["Departure Delay in Minutes"],
    df["Arrival Delay in Minutes"]
)

print("Correlation =", corr)
print("P-value =", p_value)
```

```python
if p_value < 0.05:
    print("Significant relationship exists")

else:
    print("No significant relationship exists")
```
